```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = px.data.iris()
df['e'] = df['sepal_width']/100
fig = px.scatter(
    df,
    x='sepal_width', y='sepal_length',
    color='species',
    error_x='e', error_y='e'
)
fig
```

```python
df['e_plus'] = df['sepal_width']/100
df['e_minus'] = df['sepal_width']/40
fig = px.scatter(
    df,
    x='sepal_width', y='sepal_length',
    color='species',
    error_y='e_plus', error_x='e_minus'
)
fig
```

```python
fig = go.Figure(go.Scatter(
    x=[0, 1, 2],
    y=[6, 10, 2],
    error_y=dict(
        type='data',
        array=[1, 2, 3],
        visible=True
    )
))
fig
```

```python
fig = go.Figure(go.Scatter(
    x=[1, 2, 3, 4],
    y=[2, 1, 3, 4],
    error_y=dict(
        type='data',
        symmetric=False,
        array=[0.1, 0.2, 0.1, 0.1],
        arrayminus=[0.2, 0.4, 1, 0.2]
    )
))
fig
```

```python
go.Figure(go.Scatter(
    x=[0, 1, 2],
    y=[6, 10, 2],
    error_y=dict(
        type='percent',
        value=50
    )
))
```

```python
go.Figure(go.Scatter(
    x=[1, 2, 3, 4],
    y=[2, 1, 3, 4],
    error_y=dict(
        type='percent',
        value=15,
        valueminus=25)
))
```

```python
go.Figure(go.Scatter(
    x=[1, 2, 3, 4],
    y=[2, 1, 3, 4],
    error_x=dict(
        type='percent',
        value=15
    )
))
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=['Trial 1', 'Trial 2', 'Trial 3'], y=[3, 6, 4],
    error_y=dict(
        type='data',
        array=[1, 0.5, 1.5]
    )
))

fig.add_trace(go.Bar(
    x=['Trial 1', 'Trial 2', 'Trial 3'], y=[4, 7, 3],
    error_y=dict(
        type='data',
        array=[0.5, 1, 2]
    )
))
```
